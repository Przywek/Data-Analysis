```python
def male_female_child(passenger):
    age,sex=passenger
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
def raz(passenger):
    survived,sibsp,parch=passenger
    if survived == 1:
        if sibsp==1:
            return 'Yes'
        elif parch==1:
            return 'Yes'
        else:
            return 'No'
  
```

```python
titanic_df['FamSurv'] = titanic_df[['Survived','SibSp','Parch']].apply(raz,axis=1)
```

```python
titanic_df.head()
```

```python
sns.factorplot('FamSurv',data=titanic_df,kind='count')
```

```python
def dwa(passenger):
    survived,sibsp,parch=passenger
    if survived == 1:
        if sibsp==1:
            return 'Yes'
        elif parch==1:
            return 'Yes'
        else:
            return 'No'
    else:
        return 'No'
```

```python
titanic_df['FamSurvfull']= titanic_df[['Survived','SibSp','Parch']].apply(dwa,axis=1)
```

```python
sns.factorplot('FamSurvfull',data=titanic_df,kind='count')
```
