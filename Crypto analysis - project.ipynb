```python
days=365
dt = 1/days
mu_btc = rets.mean()['BTC']
sigma_btc = rets.std()['BTC']
mu_eth = rets.mean()['ETH']
sigma_eth = rets.std()['ETH']
mu_lsk = rets.mean()['LSK']
sigma_lsk = rets.std()['LSK']
mu_game = rets.mean()['GAME']
sigma_game = rets.std()['GAME']
mu_ltc = rets.mean()['LTC']
sigma_ltc = rets.std()['LTC']
```

```python
def stock_monte_carlo(start_price,days,mu,sigma):
    price = np.zeros(days)
    price[0]= start_price
    shock= np.zeros(days)
    drift= np.zeros(days)
    for x in range(1,days):
        shock[x] = np.random.normal(loc=mu*dt,scale=sigma*np.sqrt(dt))
        drift[x] = mu*dt
        price[x] = price[x-1] + (price[x-1]*(drift[x]+shock[x]))
    return price
```

```python
st_pr_btc=13657.2
st_pr_eth=772.64
st_pr_lsk=21.13
st_pr_game=4.11
st_pr_ltc=229.03
```

```python
runs = 1000
sim_btc = np.zeros(runs)
sim_eth = np.zeros(runs)
sim_lsk = np.zeros(runs)
sim_game = np.zeros(runs)
sim_ltc = np.zeros(runs)
for run in range(runs):
    sim_btc[run]= stock_monte_carlo(st_pr_btc,days,mu_btc,sigma_btc)[days-1]
    sim_eth[run]= stock_monte_carlo(st_pr_eth,days,mu_eth,sigma_eth)[days-1]
    sim_lsk[run]= stock_monte_carlo(st_pr_lsk,days,mu_lsk,sigma_lsk)[days-1]
    sim_game[run]= stock_monte_carlo(st_pr_game,days,mu_game,sigma_game)[days-1]
    sim_ltc[run]= stock_monte_carlo(st_pr_ltc,days,mu_ltc,sigma_ltc)[days-1]
```

```python
q = np.percentile(sim_btc,1)
plt.figure(figsize=(10,5))
plt.hist(sim_btc,bins=200)
plt.figtext(-0.12,0.8,s="Start price: $%.2f" %st_pr_btc)
plt.figtext(-0.12,0.7,"Mean final price: $%.2f" % sim_btc.mean())
plt.figtext(-0.12,0.6,"VaR(0.99): $%.2f" % (st_pr_btc - q,))
plt.figtext(-0.12,0.5,"VaR(0.99): $%.2f" % q)
plt.axvline(x=q,linewidth=4,color='r')
plt.title(u"Final price distribution for Bitcoin after %s days" % days,weight='bold');
plt.show()
```
